```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib import cm 
%matplotlib inline
```

```python
road_img = cv2.imread("scene_img.jpg")
```

```python
plt.imshow(road_img)
```

```python
road_copy = road_img.copy()
road_copy.shape
```

```python
marker_img = np.zeros(road_copy.shape[:2],dtype = np.int32)
plt.imshow(marker_img)
```

```python
segments = np.zeros(road_copy.shape, dtype = np.uint8)
plt.imshow(segments)
```

```python
colors = []
for i in range(10):
    colors.append(tuple(np.array(cm.tab10(i)[:3])*255))
    
print(colors)
len(colors)
```

```python
## Global variables 
# color choice 
n_marker = 10
current_marker = 1
marks_updates = False 


```

```python
# call back function 
def mouse_callback(event,x,y,flags,param):
    global marks_updates 
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(marker_img,(x,y),10,(current_marker),-1)
        # User sees on the road image
        cv2.circle(road_copy,(x,y),10,colors[current_marker],-1)
        
        marks_updates = True
        
        
        
```

```python
# While true 
cv2.namedWindow('Road Image')
cv2.setMouseCallback('Road Image',mouse_callback)

while True:
    cv2.imshow("Watershed segments",segments)
    cv2.imshow("Road Image",road_copy)
    
    # CLOSE ALL WINDOWS 
    k = cv2.waitKey(1)
    if k == 27:
        break 
        
    # Clearing all the colors while clicking c 
    elif k == ord('c'):
        road_copy = road_img.copy()
        marker_img = np.zeros(road_img.shape[:2],dtype = np.int32)
        segments = np.zeros(road_img.shape, dtype = np.uint8)
        
    
    # UPDATE COLOR CHOICE 
    elif k > 0 and chr(k).isdigit():
        current_marker = int(chr(k))
        
    # update the markings: 
    
    if marks_updates == True:
        marker_img_copy = marker_img.copy()
        cv2.watershed(road_img, marker_img_copy)
        segments = np.zeros(road_img.shape, dtype = np.uint8)
        
        for color_ind in range(n_marker):
            segments[marker_img_copy == (color_ind)] = colors[color_ind]
            
            
            
cv2.destroyAllWindows()
        
        
    
```

## Face dtection 

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
img_1 = cv2.imread("Nadia_Murad.jpg",0)
img_2 = cv2.imread("solvay_conference.jpg",0)
img_3 = cv2.imread("Denis_Mukwege.jpg",0)

obj_face = cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml")


```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = obj_face.detectMultiScale(face_img,scaleFactor = 1.2, minNeighbors = 5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),15)
        
        
        return face_img
        
    
```

```python
result = detect_face(img_1)
plt.imshow(result,cmap="gray")
```

```python
result = detect_face(img_2)
plt.imshow(result,cmap="gray")
```

```python
result = detect_face(img_3) 
plt.imshow(result,cmap="gray")
```
